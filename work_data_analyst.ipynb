```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
Order_report = pd.read_excel('/content/Company X - Order Report.xlsx')
pincode_zones = pd.read_excel('/content/Company X - Pincode Zones.xlsx')
SKU =pd.read_excel('/content/Company X - SKU Master.xlsx')
Invoice= pd.read_excel('/content/Courier Company - Invoice.xlsx')
Rates = pd.read_excel('/content/Courier Company - Rates.xlsx')




```

```python
Order_report
Company_X_df = pd.merge(Order_report,SKU,on='SKU')
```

```python
Company_X_df['Weight_calculated'] = Company_X_df['Order Qty'] * Company_X_df['Weight (g)']

Data_1 = Company_X_df.groupby('ExternOrderNo')['Weight_calculated'].sum().reset_index()
Data_1['Weight_calculated'] = Data_1['Weight_calculated'].astype(int)
Data_1['Weight (kg)'] = Data_1['Weight_calculated'] / 1000
Data_1['Weight (kg)']= Data_1['Weight (kg)'].round(decimals = 1)
Data_1
```

```python
Data_1 = Data_1.rename(columns={'ExternOrderNo':'Order ID','Weight (kg)':'Total weight as per X (KG)'})
Data_1
```

```python
import math

def round_up_to_nearest_half_int(num):
    return math.ceil(num * 2) / 2
```

```python
Data_1['Weight slab as per X (KG)'] = Data_1['Total weight as per X (KG)'].apply(round_up_to_nearest_half_int)
```

```python
def sample_function(row):
    cost=0


    counter = row['Weight_slab_metric']
    if counter == 1 & bool(row['Type of Shipment']=='Forward charges'):

        cost = row['fwd_fixed']
        status = bool(row['Type of Shipment']=='Forward charges')
        return cost


    elif counter == 1 &  bool(row['Type of Shipment']=='Forward and RTO charges'):

        cost = row['rto_fixed']
        return cost

    elif counter > 1 & bool(row['Type of Shipment']=='Forward charges'):

        cost_a=0
        cost_b=0
        final_cost=0
        cost_a = row['fwd_fixed']
        cost_b = row['fwd_additional'] * (counter-1)
        final_cost= cost_a + cost_b
        return final_cost

    else:

        cost_a=0
        cost_b=0
        final_cost=0
        cost_a = row['rto_fixed']
        cost_b = row['rto_additional'] * (counter-1)
        final_cost = cost_a + cost_b
        return final_cost

```

```python
Data_4['Weight_slab_metric'] = Data_4['Weight slab as per X (KG)']/0.5

Data_4['Expected Charge as per X (Rs.)'] =  Data_4.apply(sample_function,axis=1)

Data_4
```
